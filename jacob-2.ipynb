```python
library_exp_counts = df.groupby("group")[["beautiful_soup", "flask", "matplotlib", "nltk",  "numpy", "pandas", "pytorch"]].sum()
library_exp_counts
```

```python
def df_to_latex(df):
    sep = ""
    for feat in df.columns:
        print(r" & \texttt{" + feat.replace("_", r"\_") + "}", end = "")
    print(r" \\")
    for n, row in df.iterrows():
        print(n, end="")
        for col in row:
            print(" &", col, end = "")
        print(r" \\")

df_to_latex(library_exp_counts)
```

```python
yrs_prog_numerical = df["yrs_prog"].replace(["<1 year ago", "Less than 1 year ago"], 0).replace("10 or more years ago", 10).astype(int)
yrs_prog_numerical.median()
```

```python
yrs_prog_weekly_numerical = df["yrs_prog_weekly"].replace(["<1 year", "Less than 1 year"], 0).replace("10 or more years", 10).astype(int)
yrs_prog_weekly_numerical.median()
```

```python
demog_numerical = df[["group", "self_est_prog_exp"]].copy().rename(columns={"self_est_prog_exp" : "self-estimated programming experience"})
demog_numerical["years programming experience"] = yrs_prog_numerical
demog_numerical["years programming on a weekly basis"] = yrs_prog_weekly_numerical
demog_numerical
```

```python
demog_by_group = demog_numerical.groupby("group").median()
df_to_latex(demog_by_group)
```
